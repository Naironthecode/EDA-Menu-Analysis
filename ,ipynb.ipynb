```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/Darsh/Downloads/menu.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])

sales_trend = df.groupby('Date')['Sales'].sum()

sales_trend.plot(figsize=(10,5))
plt.title("Sales Over Time")
plt.show()
```

```python
df.columns
```

```python
sns.barplot(x='Category', y='Calories', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.title("Analysis of Menu Items")
plt.xlabel("Category")
plt.ylabel("Calories")
```

```python
sns.barplot(x='Category', y='Calories', data=df)

plt.title("Analysis of Menu Items")
plt.xlabel("Category")
plt.ylabel("Calories")
plt.xticks(rotation=45)

plt.show()
```

```python
top5 = df.sort_values(by='Calories', ascending=False).head(5)
top5
```

```python
low5 = df.sort_values(by='Calories').head(5)
low5
```

```python
category_avg = df.groupby('Category')['Calories'].mean().sort_values(ascending=False)
category_avg
```

```python
plt.figure(figsize=(12,6))
category_avg.plot(kind='bar')

plt.title("Average Calories per Category")
plt.xlabel("Category")
plt.ylabel("Calories")

plt.xticks(rotation=45)
plt.show()
```

```python
corr = df.corr(numeric_only=True)
corr
```

```python
plt.figure(figsize=(14,10))

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")

plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.title("Correlation Heatmap")
plt.show()
```

```python
important_cols = ['Calories', 'Total Fat', 'Sodium', 'Sugars', 'Protein']

corr_small = df[important_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_small, annot=True, cmap='coolwarm')

plt.title("Important Features Correlation")
plt.show()
```
